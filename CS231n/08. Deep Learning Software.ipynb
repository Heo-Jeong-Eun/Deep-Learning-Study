## CPU vs GPU
##### 컴퓨터의 뇌와 같은 CPU(Central Processing Unit)는 기억, 해석, 연산, 제어를 수행한다.

###### CPU는 적은 코어수로 연속적인 작업을 처리한다. 반면 컴퓨터 그래픽 rendering 하기 위해 GPU(Graphics Processing Unit)는 많은 코어로 병렬로 수행하고, 내부적으로 RAM을 사용한다. 또 GPU는 상호작용을 생각해줘야 한다. 데이터를 읽을 때 종종 Bottleneck이 발생한다.

## Bottleneck
##### 병목현상은 전체 시스템의 성능이나 용량이 하나의 구성 요소로 인해 제한을 받는 현상이다.
##### 한 번에 처리할 수 있는 데이터의 양보다 처리할 수 있는 능력이 충분하지 않을 경우 발생한다.

## Pytorch
##### pytorch는 3가지 abstraction이 있다.


*   tensor : 다차원배열로 GPU에서 run 될 수 있도록 하는 일종의 numpy 배열
*   variable : computational graph의 노드 중 데이터와 gradient
*   module : neural network의 layer, state와 학습 가능한 weight

##### pytorch는 tensorflow와 다르게 graph가 dynamic 하다. 또한 forward pass 시 그래프를 다시 구성하여 학습시킬때마다 새로운 그래프를 그린다. 이외에도 optimizer를 이용해 learning rate를 쉽게 설정할 수 있다.
